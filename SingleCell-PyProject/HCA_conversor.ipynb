```python
import json

with open('hits_raw.json') as f:
    hits = json.load(f)
```

```python
def hca2ont(text):
    if text is None:
        return None
    
    if type(text) is list:
        return list(map(hca2ont, text))
    
    aux = list(text.title())
    
    for i in range(len(text)):
        if text[i].isupper():
            aux[i] = text[i]
    
    return ''.join(aux).replace(' ', '')
```
