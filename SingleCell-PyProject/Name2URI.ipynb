```python
import requests
from lxml import html
```

```python
keyword = "neuron"
ontology = "CL"

seed_url = f"http://www.ontobee.org/search?ontology={ontology}&keywords={keyword}&submit=Search+terms"
```

```python
# User-Agent
headers = {
    "user-agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.83 Safari/537.36"
}
```

```python
answer = requests.get(seed_url, headers=headers)
```

```python
def parse_name(li):
    name = li.xpath(".//text()")
    name = ''.join(name)
    name = name.strip()
    name = name.split(':')[0]
    
    return name
```

```python
parser = html.fromstring(answer.text)

terms = parser.xpath("//li[@class='search-list']")

terms_processed = []

for term in terms:
    
    URI = term.xpath("./a/text()")[0]
    lis = term.xpath("./ul/li")
    
    names = []
    
    for li in lis:
        name = parse_name(li)
        names += [name]

    terms_processed.append({
        'name': name,
        'URI': URI
    })
    
terms_processed
```
