```python
path = '../SingleCell-Files/percentiles/E-MTAB-4547.percentiles.csv'
df = pd.read_csv(path)
```

```python
def df_to_dict(df, project_id):
    df_dict = {
        'project_id': project_id,
        'genes': []
    }
    for _, row in df.iterrows():
        
        gene_dict = {
            'name': row['gen_name'],
            'cell_types': []
        }        
                
        for name, percentile in row[1:].items():
            gene_dict['cell_types'].append({
                'name': name,
                'percentile': percentile
            })
        
        df_dict['genes'].append(gene_dict)
                
    return df_dict
```

```python
dict_df = df_to_dict(df, 'E-MTAB-4547')
```

```python
requests.post(url, json=dict_df)
```

```python
server_name = 'http://194.4.103.244:3030'
service_name = 'ds'
request_url = server_name + '/' + service_name
```

```python
query = '''
    PREFIX a: <http://www.semanticweb.org/alicia/ontologies/2020/8/singleCellRepositories#>
    PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
    SELECT 
    ?project_ID
    WHERE
    {
      ?project rdf:type a:Project ;
               a:PR.hasProjectID ?project_ID .
    }
'''

response = requests.post(request_url, 
                         data={'query': query})
```
